```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
di = pd.read_excel('/content/drive/MyDrive/Estadistica Industria/C3/datos/datos_industrias.xlsx')
di.head()
```

```python
# prompt: necesito un analisis de regresión lineal simple que la variable X=Salario y Y=Tasa_Crecimiento

# Import necessary libraries (if not already imported)
import statsmodels.api as sm

# Prepare the data
X = di['Salario']
y = di['Tasa_Crecimiento']

# Add a constant to the independent variable (required for statsmodels)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())

# Make predictions
predictions = model.predict(X)

# Plot the regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Salario', y='Tasa_Crecimiento', data=di)
plt.title('Regression Line: Salary vs. Growth Rate')
plt.xlabel('Salary')
plt.ylabel('Growth Rate')
plt.show()
```
